```python
# define function for checking all the cells which match up to obtain a sum
def findsum(barcode):
    # Take count for number of times of gene appearance in cell
    genefreq = 0
    # Skip the total lines and comments
    skip = 0
    mtx_file_other  = open('C:/Users/mastercljohnson/Desktop/Assignment/M44_matrix.mtx', 'r')
    for bar in mtx_file_other:
        # Check if row is a comment or a total
        if ( skip == 1 ):
            skip = 0
            continue
            
        if ( bar.find("%%") != -1 ):
            continue
            
        if ( bar.find("%") != -1):
            skip = 1
            continue
            
        # Change the row into an array
        bar = bar.replace('\n', '')
        array = bar.split(' ')
        
        if ( str(barcode) == array[1] ):
            if (int(array[2]) > 0):
                genefreq+= int(array[2])
                print("bar matched")
                
    mtx_file_other.close()
        
    return genefreq


# Read interest list files from io
interestlist  = open('C:/Users/mastercljohnson/Desktop/Assignment/365_EMT_gene_signatures.txt', 'r')



# Write out text files for number of cells for each gene of interest
cellspergene = open('C:/Users/mastercljohnson/Desktop/Assignment/CellPerGene.txt','a')

i = 0
for line in interestlist:
    print(line)
    matchindex = -1
    index = 0
    # Recreating the generator because iterators only can iterate once
    genes  = open('C:/Users/mastercljohnson/Desktop/Assignment/M44_genes.tsv', 'r')
    for gene in genes:
        # Parse the strings in the gene list
        gene = gene.upper()
        row_arr = gene.split('\t')
        row_arr[1].replace('\n','')
        
        # Check if the row of gene matches with intended gene
        if ( line == row_arr[1] ):
            # rows are always displaced by one
            matchindex = index + 1
            print("gene matched")
        
            
        # Increment index
        index+=1
    # Close file    
    genes.close()
    
    # Checks if gene exists in list of genes
    if ( matchindex == -1 ):
        continue
    # Take count for number of cells
    count = 0
    # Check for nonzero cells
    skip = 0
    mtx_file  = open('C:/Users/mastercljohnson/Desktop/Assignment/M44_matrix.mtx', 'r')
    for row in mtx_file:      
        # Check if row is a comment or a total
        
        if ( skip == 1 ):
            skip = 0
            continue
            
        if ( row.find("%%") != -1 ):
            continue
            
        if ( row.find("%") != -1):
            skip = 1
            continue
            
        # Change the row into an array
        row = row.replace('\n', '')
        array = row.split(' ')
        
        
        # Increment cell number if greater than 0 and found for number of cells per gene
        # Also append the cell into the 
        if ( str(matchindex) == array[0] ):
            if ( int(array[2]) > 0):
                count+=1
                
                # Only do this for one file for the sake of time and demonstration
                # as specified in the prompt
                print(i)
                if ( i == 0 ):

                    # Show the cell barcode in the file
                    barindex = int(array[1])
                    barsum = findsum(barindex)
                
                    cell_barcodes  = open('C:/Users/mastercljohnson/Desktop/Assignment/M44_barcodes.tsv', 'r')
                
                    # using a string as an intermediate to avoid changing file
                    string = cell_barcodes.readlines()[barindex-1]
                
                    cell_barcodes.close()
                
                    string = string.replace('\n','')
                
                
                    # Write out text file for cells for each gene of interest
                    filename = "C:/Users/mastercljohnson/Desktop/Assignment/Barcodes" + line.replace('\n','')+".txt"
                    newfile = open(filename,'a')
                    # barindex - 1 is for the displacement for row 1 being index 0
                    newfile.write(string +"\t" + str(barsum))
                    print("wrote cell file")
                
                    # Close file for each list of interest
                    newfile.close()

                
    # Checks if gene is found in more than one cell
    if ( count > 0 ):
        cellspergene.write(line.replace('\n','') + "\t" + str(count) + "\n")
        print("write into cell count")
        
    
    mtx_file.close()
    
    i+=1
    
    
cellspergene.close()

interestlist.close()

print("DONE")

```
