```python
# Create a hash table for cells and for matching genes
genesdict = {}

# Dictionary/Hashtable for number of cells each gene has
genecountdict = {}

# Dictionary for holding cells which a gene is contained in
cellsdict = {}

# Dictionary holds the number of genes in each cell
cellscountdict = {}

# Read interest list files from io
interestlist  = open('C:/Users/mastercljohnson/Desktop/Assignment/365_EMT_gene_signatures.txt', 'r')
mtx_file  = open('C:/Users/mastercljohnson/Desktop/Assignment/M44_matrix.mtx', 'r')

print("passed")

for line in interestlist:
    index = 0
    # Recreating the generator because iterators only can iterate once
    genes  = open('C:/Users/mastercljohnson/Desktop/Assignment/M44_genes.tsv', 'r')
    for gene in genes:
        # Parse the strings in the gene list
        gene = gene.upper()
        row_arr = gene.split('\t')
        row_arr[1].replace('\n','')
        
        # Check if the row of gene matches with intended gene
        if ( line == row_arr[1] ):
            # rows are always displaced by one
            genesdict[index + 1] = line
        
            
        # Increment index
        index += 1
    # Close file    
    genes.close()
    
print("genes matched")

# Used as an intermediate to skip 
skip = 0
for row in mtx_file:      
    # Check if row is a comment or a total
        
    if ( skip == 1 ):
        skip = 0
        continue
            
    if ( row.find("%%") != -1 ):
        continue
            
    if ( row.find("%") != -1):
        skip = 1
        continue
            
    # Change the row into an array
    row = row.replace('\n', '')
    array = row.split(' ')
        
        
    # Increment cell number if greater than 0 and found for number of cells per gene
    # Use dictionary mapping
    if int(array[0]) in genesdict.keys() :
        if ( int(array[2]) > 0):
            # This is for gene count
            if int(array[0]) in genecountdict.keys():
                genecountdict[int(array[0])] = genecountdict.get(int(array[0])) + 1
            else:
                genecountdict[int(array[0])] = 1
            
            # This is for cell contained and cell count
            if int(array[0]) in cellsdict.keys():
                cellsdict[int(array[0])] = cellsdict[int(array[0])]+[int(array[1])]
            else:
                cellsdict[int(array[0])] = [int(array[1])]
                
                
    # This is for finding the sum of all genes for a particular cell
    if int(array[1]) in cellscountdict.keys() :
        if ( int(array[2]) > 0):
            cellscountdict[int(array[1])] = cellscountdict.get(int(array[1])) + int(array[2])
    else:
        if ( int(array[2]) > 0):
            cellscountdict[int(array[1])] = int(array[2])
            
print("Start to Write")
            
# Write out text files for number of cells for each gene of interest
cellspergene = open('C:/Users/mastercljohnson/Desktop/Assignment/HashCellPerGene.txt','a')

for key in genesdict.keys():
    # Write to file
    writestring = str(genesdict[key]).replace('\n','')
    #print (writestring)
    if key in genecountdict.keys():
        cellspergene.write( writestring + "\t" + str(genecountdict[key]) +"\n")
   
        # For each gene of interest write for each cell total
        filename = "C:/Users/mastercljohnson/Desktop/Assignment/BarcodesHash" + str(genesdict[key]).replace('\n','') +".txt"
        newfile = open(filename,'a')
    
        for bar in cellsdict[key]:    
            # Read in barcode lines
            cell_barcodes  = open('C:/Users/mastercljohnson/Desktop/Assignment/M44_barcodes.tsv', 'r')
            # Extract wanted line
            string = cell_barcodes.readlines()[bar-1]
            # Close file
            cell_barcodes.close()
        
            string = string.replace('\n','')
            string = string + "\t" + str(cellscountdict[bar] )
            newfile.write(string)
        
        
        # Close file    
        newfile.close()
    
#Close file
cellspergene.close()
print("Done")
```
